## sort into time slots

```python
#time_slot_0 : 使用者在 01:00:00 ~ 09:00:00 間有無活動（機率）。
#time_slot_1 : 使用者在 09:00:00 ~ 17:00:00 間有無活動（機率）。
#time_slot_2 : 使用者在 17:00:00 ~ 21:00:00 間有無活動（機率）。
#time_slot_3 : 使用者在 21:00:00 ~ 01:00:00 間有無活動（機率）。
```

```python
from datetime import datetime, time

def is_time_between(begin_time, end_time, check_time=None):
    if begin_time < end_time:
        return int(check_time >= begin_time and check_time <= end_time)
    else: # crosses midnight
        return int(check_time >= begin_time or check_time <= end_time) #return int of boolean
    
# testing for activity in time period 1:00 ~ 9:00
for d in data_1092.time.iloc[2495:2500]:
    print (is_time_between(time(1,00), time(9,00), d))

data_1092.time.iloc[2495:2500]
```

```python
#time_slot_0 : 使用者在 01:00:00 ~ 09:00:00 間有無活動（機率）。
#time_slot_1 : 使用者在 09:00:00 ~ 17:00:00 間有無活動（機率）。
#time_slot_2 : 使用者在 17:00:00 ~ 21:00:00 間有無活動（機率）。
#time_slot_3 : 使用者在 21:00:00 ~ 01:00:00 間有無活動（機率）。

a = []
b = []
c = []
e = []
for d in data_1092.time:
    a.append(is_time_between(time(1,00), time(9,00), d))
    b.append(is_time_between(time(9,00), time(17,00), d))
    c.append(is_time_between(time(17,00), time(21,00), d))
    e.append(is_time_between(time(21,00), time(1,00), d))
```

```python
data_1092['1_9'] = a
data_1092['9_17'] = b
data_1092['17_21'] = c
data_1092['21_1'] = e
```
